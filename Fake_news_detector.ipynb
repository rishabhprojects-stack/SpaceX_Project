```python
real_news['label'] = 1
fake_news['label'] = 0
```

```python
combined_news.drop(['title', 'subject', 'date'], axis = 1, inplace = True)
```

```python
shuffled_news = combined_news.sample(frac = 1) # Shuffled dataset, so that the labels appear randomly
```

```python
shuffled_news.head()
```

```python
shuffled_news.reset_index(inplace = True)
```

```python
shuffled_news.drop(['index'], axis = 1, inplace = True)
```

```python
shuffled_news.head()
```

```python
import re
```

```python
def wordopt(text):
    text = text.lower() # Make text lowercase
    text = re.sub(r'https?://\S+|www\.\S+','', text) # Remove URLs
    text = re.sub(r'<.*?>','',text) # Remove HTML tags
    text = re.sub(r'[^\w\s]','',text) # Remove punctuation
    text = re.sub(r'\d','',text) # Remove digits
    text = re.sub(r'\n',' ',text) # Remove newline characters
    return text
```

```python
shuffled_news['text'] = shuffled_news['text'].apply(wordopt)
```

```python
shuffled_news.head()
```

```python
x, y = shuffled_news['text'], shuffled_news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
def model(news):
    testing_news = {'text' : [news] }
    model_test = pd.DataFrame(testing_news)
    model_test['text'] = model_test['text'].apply(wordopt)
    model_x_test = model_test['text']
    model_xv_test = vectorization.transform(model_x_test) # Assuming 'vectorization' is the function object
    predict_lr, predict_dtc, predict_rfc, predict_gbc = LR.predict(model_xv_test),DTC.predict(model_xv_test),\
    rfc.predict(model_xv_test),gbc.predict(model_xv_test)
    return 'LR Prediction: {} \nGBC Prediction: {} \nDTC Prediction: {} \nRFC Prediction: {}'.format(
        result(predict_lr[0]), result(predict_gbc[0]), result(predict_dtc[0]), result(predict_rfc[0]))
```
